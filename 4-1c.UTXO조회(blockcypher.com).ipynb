```python
import requests


def baseurl(addr):
    if addr.startswith('1'):
        return 'https://api.blockcypher.com/v1/btc/main/'
    elif addr.startswith('m'):
        return 'https://api.blockcypher.com/v1/btc/test3/'

    
def balance(addr):
    resp = requests.get(baseurl(addr) + f'addrs/{addr}/balance')
    data = resp.json()
    return data['balance']


def print_utxo(addr):
    resp = requests.get(baseurl(addr) + f'addrs/{addr}?unspentonly=true')
    if resp.status_code != 200:
        print("\n", resp.text)
        return
    
    data = resp.json()

    for utxo in [tx for tx in data['txrefs'] if tx['tx_output_n'] > 0]:
        print("tx_hash = ", utxo['tx_hash'])
        print("tx_output_n = ", utxo['tx_output_n'])
        print("value = ", utxo['value'])
        print()

    print("Total Balance = ", data['balance'], '(Satoshi)')
```

```python
print(balance('1Nh7uHdvY6fNwtQtM1G5EZAFPLC33B59rB'))
```

```python
print_utxo('1Nh7uHdvY6fNwtQtM1G5EZAFPLC33B59rB')
```
